```python
def get_html(url=''):
    return urlopen(url).read().decode('utf-8')
```

```python
import re

def find_words_blocks(html):
    return re.findall("<b>.*?</b>", html, flags=re.DOTALL)
```

```python
def get_words(block):
    return re.sub("<.*?>|(.*?Compartilhar.*?via.*?WhatsApp.*)|(.*?Aizeta.*)", '', block, flags=re.DOTALL)
```

```python
def write_file(file_name="/scratch/PUC/knime-workspace/metaflows_professor/1-SimilaritySearch/listas_pos_tagger/substantivos.txt", content=''):
    arq = open(file_name, 'w')
    arq.write('{}\n'.format(content))
    arq.close()
```

```python
from urllib.request import urlopen
import time
from random import randint

a = ord('a')
alphabet = [chr(i) for i in range(a,a+26)]
main_url = "https://dicionario.aizeta.com/verbetes/substantivo/{letter}/{page}"

for letter in alphabet:
    print('\nLetter', letter)
    words = set()
    page = 1
    print (page, end=', ')
    
    while True:
        html = get_html(url=main_url.format(letter=letter, page=page))
        
        blocks = find_words_blocks(html)

        blocks_to_remove = []
        for block in blocks:
            word = get_words(block)
            if len(word) > 1:
                words.add(word.lower())
            else:
                blocks_to_remove.append(block)
                
        for block in blocks_to_remove:
            blocks.remove(block)
        
        if len(blocks) == 0:
            break

        time.sleep(randint(1, 10))
        
        page += 1
        print (page, end=', ')

    print ('Number of words: ', len(words))
    for word in words:
        write_file(content=word)
```
