```python
from urllib.request import urlopen
def get_html(url=''):
    try:
        return urlopen(url).read().decode('latin1')#.encode('utf-8')
    except:
        try:
            return urlopen(url).read().decode('utf-8')#.encode('latin1')
        except:
            try:
                return urlopen(url).read().decode('ISO-8859-1')#.encode('utf-8')
            except:
                try:
                    return urlopen(url).read().encode('utf-8')#.encode('utf-8')
                except:
                    return "\t\t - ERROR - {}".format(url)
```

```python
import re

def find_verbs_links(html):
    return re.findall(u"\/verbo\-.*?\/", html, flags=re.DOTALL)
```

```python
def find_conjugated_verbs(html):
    return re.findall(u"<span\s+class\s*\=\s*\"f\"\s*>.*?<\s*\/span\s*>", html, flags=re.DOTALL)
```

```python
def get_words(block):
    return re.sub("<.*?>", '', block, flags=re.DOTALL)
```

```python
def write_file(file_name="", content_list=[]):
    arq = open(file_name, 'a')
    for content in content_list:
        arq.write('{}\n'.format(content))
    arq.close()
```

```python
import time
from random import randint

main_url = u"https://www.conjugacao.com.br/verbos-populares/{page}"
page = 1
count_verbs = 0
verbs_links = set()

while True:
    print (main_url.format(page=page))
    
    html = get_html(url=main_url.format(page=page))

    verbs = find_verbs_links(html)
    
    for verb in verbs:
        verbs_links.add('https://www.conjugacao.com.br{verb}'.format(verb=verb))
    
    count_verbs += len(verbs_links)
    if len(verbs) == 0:
        break
        
    time.sleep(randint(1, 10))
    
    page += 1

print("Count verbs: ", len(verbs_links))

write_file(content_list=list(verbs_links))
        
```
